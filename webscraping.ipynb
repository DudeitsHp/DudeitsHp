```python
import pandas as pd 
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
```

```python

req = Request("https://docs.google.com/spreadsheets/d/e/2PACX-1vSrRVXgdT8sAPInyPhAQ7b3okwDgi6u66sGdt-vZy6WdLjaqsMq7I4MXdO5fLFwmQ/pubhtml")
html_page = urlopen(req)
```

```python
soup = BeautifulSoup(html_page, 'html.parser')
```

```python
links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))
```

```python
print(links)
```

```python
df = pd.DataFrame(links)
```

```python
df.to_csv('data/resumeurl.csv', index=False)
```
